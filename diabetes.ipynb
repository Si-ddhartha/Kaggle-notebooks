```python
# There is a lot of imbalancement
# so we will use weighted loss function
# to address class imbalance
```

```python
# Some preprocessing

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = data.drop('Diabetes_012', axis=1)
y = data['Diabetes_012'].values
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Compute class weights
from sklearn.utils.class_weight import compute_class_weight

class_weight = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train.numpy())
class_weight = torch.tensor(class_weight, dtype=torch.float32)
```

```python
from torch.utils.data import TensorDataset, DataLoader

train_dataset = TensorDataset(X_train, y_train)
test_dataset = TensorDataset(X_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=True)
```

```python
# Defining the model
class DiabetesModel(nn.Module):
    
    def __init__(self):
        super(DiabetesModel, self).__init__()
        
        self.fc1 = nn.Linear(X_train.shape[1], 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 3)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        
        return x
```

```python
model = DiabetesModel()
loss_fun = nn.CrossEntropyLoss(weight=class_weight)
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Training
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = loss_fun(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
    
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader):.4f}')
```

```python
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for inputs, labels in test_loader:
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        
accuracy = 100 * (correct / total)
print(f'Accuracy on test set: {accuracy:.2f}%')
```
